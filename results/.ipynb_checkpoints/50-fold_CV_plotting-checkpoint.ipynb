```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tb = np.load("theoretical_best.npy", allow_pickle=True).item()
cv = np.load("cv_results.npy", allow_pickle=True).item()
```

```python
tb_rmse = tb["pred_rmse"]
```

```python
cv['0.1-0.9']["rmse"]
```

```python
plt.figure(figsize=(10,10))
all_splits = []
split_labels = []

for split in cv.keys():
    split_labels += [split]
    all_splits += [cv[split]["rmse"]]

plt.boxplot(all_splits)
plt.hlines(tb_rmse, 0,6, linestyles='dashed', color='r')

plt.ylabel("RMSE", fontsize=18)

plt.xticks(np.arange(1,len(all_splits)+1), split_labels, fontsize=17)
plt.ylim(0,10)
plt.show()
plt.close()
```
