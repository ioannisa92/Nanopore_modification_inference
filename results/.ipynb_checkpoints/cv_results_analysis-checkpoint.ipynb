```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
cv_res = np.load("cv_results.npy", allow_pickle=True).item()
```

```python
plt.figure(figsize=(10,10))
for key in cv_res.keys():
    
    plt.plot(cv_res[key]['train_history']['loss'], label=key)
plt.ylim(0,40)
plt.legend()
plt.title("Loss vs Epoch. Zoomed in (y-axis zoom)", fontsize=20)
plt.ylabel("Loss", fontsize=15)
plt.xlabel("Epoch", fontsize=15)
plt.show()
plt.close()
```

```python
cv_res['0.9-0.1'].keys()
```

```python
fig = plt.figure(figsize=(10,10))

ax1 = fig.add_subplot(221)
for key in cv_res.keys():
    key_ = list(map(lambda x:str(round(float(x),1)), key.split('-')))
    
    key_ = '-'.join(key_)
    ax1.plot(cv_res[key]['train_history']['loss'], label=key_)
ax1.legend()
ax1.set_title("Loss vs Epoch", fontsize=20)
ax1.set_ylabel("Loss", fontsize=15)
ax1.set_xlabel("Epoch", fontsize=15)


ax2 = fig.add_subplot(222)
for key in cv_res.keys():
    key_ = list(map(lambda x:str(round(float(x),1)), key.split('-')))
    
    key_ = '-'.join(key_)
    
    ax2.plot(cv_res[key]['r'], label=key_)
ax2.legend()
ax2.set_title("Pearson Correlation on Test", fontsize=20)
ax2.set_ylabel("Correlation Coefficient", fontsize=15)
ax2.set_xlabel("CV-5fold Folds", fontsize=15)

ax3 = fig.add_subplot(223)
for key in cv_res.keys():
    key_ = list(map(lambda x:str(round(float(x),1)), key.split('-')))
    
    key_ = '-'.join(key_)
    
    ax3.plot(cv_res[key]['r2'], label=key_)
ax3.legend()
ax3.set_ylim(0,1.2)
ax3.set_title("R2 on Test", fontsize=20)
ax3.set_ylabel("R2", fontsize=15)
ax3.set_xlabel("CV-5fold Folds", fontsize=15)

ax4 = fig.add_subplot(224)
for key in cv_res.keys():
    key_ = list(map(lambda x:str(round(float(x),1)), key.split('-')))
    
    key_ = '-'.join(key_)
    
    ax4.plot(cv_res[key]['rmse'], label=key_)
ax4.legend()
ax4.set_title("RMSE on Test", fontsize=20)
ax4.set_ylabel("RMSE", fontsize=15)
ax4.set_xlabel("CV-5fold Folds", fontsize=15)
plt.tight_layout()
```
